```python
def max_divisor(n):
  maximo_actual = 0
  i = 1
  while i < n:
    if n % i == 0:
      maximo_actual = i
    i += 1
  return maximo_actual
max_divisor(8)
```

```python
def potencia_positiva(base, exponente):
  if exponente == 0:
    return 1
  else:
    resultado = 1
    while exponente > 0:
      resultado *= base
      exponente -= 1
    return resultado
```

```python
def cuenta(tope):
  i = 0
  while i < tope:
    print(i)
    i += 1

cuenta (3)
```

```python
def calculo(numero):
  resultado = (numero - 3) ** 3
  return resultado
salida = calculo(5)
```
